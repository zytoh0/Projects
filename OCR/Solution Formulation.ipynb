Import packages

```python
import matplotlib.pyplot as plt, cv2, pytesseract
```

Load image file

```python
# image file path
path='./sampleCaptchas/input/input00.jpg'
img = cv2.imread(path,0)
```

Exploratory Data Analysis (EDA)

```python
plt.imshow(img);
```

```python
# histogram of image to find a suitable threshold
plt.hist(img.ravel(),256,[0,256])
plt.show()
```

Data pre-processing

```python
# setting threshold
threshold = 100
#thresholding
def thresholding(image):
    return cv2.threshold(image, threshold, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]
img = thresholding(img)
```

Inference

```python
plt.imshow(img);
plt.title(f'Inference: {pytesseract.image_to_string(img)}');
```
